<h2><b>12-Solicitudes a la web http</b></h2>

```python
import requests

response = requests.get('https://www.platzi.com')
print(dir(response))
```

```python
print(response.status_code) #Responde 200 que significa ok
```

```python
print(response.headers) #Imprime el headers en tipo diccionario
```

```python
print(response.headers['Access-Control-Allow-Methods']) #Accediendo a uno de los valores del diccionario
```

```python
print(response.text) #respuesta en texto, es la página web
```

<h2><b>13-Como trabajar un documento HTML</b></h2>

```python
import bs4 #importamos beatiful soup 
#(ayuda a analizar gramatical/ el dcto html, utiliza los selectores css... 
#convierte el dcto html en una estructura donde podemos generar querys)

soup = bs4.BeautifulSoup(response.text, 'html.parser') #linea para la conversión y poder generar querys
print(f'Titulo de la página: {soup.title.text}')
print("*********")
print("Lista de nodos: {}".format(soup.select('meta[name=description]')))#accedemos al selector meta, vemos que nos regresa una lista de nodos
print("*********")
print('Etiqueta content = '+soup.select('meta[name=description]')[0]['content']) #accedemos al primer valor de la lista
```

```python
response2 = requests.get('https://www.platzi.com/cursos')
soup2 = bs4.BeautifulSoup(response2.text, 'html.parser')
courses_links = soup2.select('.CourseCategories-item') #accedemos a los objetos de la clase .CourseCategories-item
courses = [course['href'] for course in courses_links] #con un list comprehesion queremos la etiqueta href por cada uno
#de los cursos obteneidos en la etiqueta
for course in courses:
    print(course)
```
